1 - Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
import tensorflow as tf
```

2 - Load Model

```python
model = tf.keras.models.load_model('audiobooks_model.h5')
```

2 - Load Dataset

```python
raw_data = np.genfromtxt('audiobooks_data_new.csv', delimiter=',')
raw_data_inputs = raw_data[:,1:]
raw_data_inputs
```

3 - Scale Data

```python
scaled_raw_data_inputs = preprocessing.scale(raw_data_inputs)
scaled_raw_data_inputs
```

4 - Prediction

```python
scaled_raw_data_outputs = np.argmax(model.predict(scaled_raw_data_inputs),1)
scaled_raw_data_outputs
```

5 - Export Result

```python
data = raw_data.copy()
data = pd.DataFrame(data)
data
```

```python
data.shape
```

```python
scaled_raw_data_outputs = pd.DataFrame(scaled_raw_data_outputs)
scaled_raw_data_outputs
```

```python
data = pd.concat([data, scaled_raw_data_outputs], axis=1)
data
```

```python
data.to_csv('audiobooks_data_new_result.csv', index=False)
```
